## Print Standard Decomposition Values

```python
import json

from experiments_acs.aggregation_refinement.standard_values.standard_value_loader import load_standard_values
```

```python
standard_values = load_standard_values()
```

```python
for key, level_decomp in standard_values["decomposition_attributes"].items():
    print(key)
    print("-----")
    # print(json.dumps(level_decomp, indent=2))
    splits = level_decomp["splits"]
    assert len(splits) == 2
    print(f"Attribute: {level_decomp['attribute']}")
    print(f"Split (left): [{", ".join([str(s) for s in sorted(splits[0])])}]")
    print(f"Split (right): [{", ".join([str(s) for s in sorted(splits[1])])}]")
    print("*" * 80)
```
