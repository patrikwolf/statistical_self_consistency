## Thresholding Answers

```python
from data_loader_global_opinion_qa.data_loader import load_goqa_data, get_question_indices_for_country
```

```python
# Load GlobalOpinionQA dataset
countries = ["Germany", "United States", "France", "Japan", "Belgium", "Italy"]
row_indices = get_question_indices_for_country(countries=countries)
survey_data = load_goqa_data(row_indices=row_indices, countries=countries)
```

```python
for idx, item in enumerate(survey_data):
    print("\n" + "*" * 80 + "\n")
    print(f"Question identifier: {item['question_identifier']}\n")
    print(f"Question:\n{item['question']}\n")
    print(f"Answer:\n{item['options']}")
```

#### Binary Outcome

```python
for idx, item in enumerate(survey_data):
    if len(item['options']) <= 3:
        print("\n" + "*" * 80 + "\n")
        print(f"Question identifier: {item['question_identifier']}\n")
        print(f"Question:\n{item['question']}\n")
        print(f"Answer:\n{item['options']}")
```

```python
good_questions = [55, 136, 345, 413, 573, 839, 1108, 1445, 1512, 1531, 1644, 2076, 2165, 2166]
# ["Q55", "Q136", "Q191", "Q295", "Q336", "Q345"]
```

#### Print Good Questions

```python
shared_countries = None
full_survey_data = load_goqa_data()
for idx in good_questions:
    item = full_survey_data[idx]
    print("\n" + "*" * 80 + "\n")
    print(f"Question identifier: {item['question_identifier']}\n")
    print(f"Question:\n{item['question']}\n")
    print(f"Answer:\n{item['options']}")
    if shared_countries is None:
        shared_countries = set(item["selections"].keys())
    else:
        shared_countries = set(set(item["selections"].keys()) & set(shared_countries))

print("------")
print(f"Shared countries: {shared_countries}")
```
